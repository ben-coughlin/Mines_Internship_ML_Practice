## 1. What is a Jupyter notebook?

A Jupyter notebook is a document made of **cells**. There are two kinds you'll use:

- **Code cells** hold Python. You *run* them and their output appears right below.
- **Markdown cells** hold formatted text (like this one). You *render* them into nice headings and lists.

Compare this to a plain `.py` script, which is one long file you run all at once from the terminal. A notebook lets you run **one small piece at a time**, see the result, then run the next piece. That "poke at it, look, poke again" loop is exactly how you explore data.

**To run a cell:** click it, then press Shift + Enter (that runs the cell and moves you to the next one). The cell just below is a code cell, so click it and press Shift + Enter now.

```python
# This is a CODE cell. The lines starting with "#" are comments (Python ignores them).
# Press Shift + Enter to run it. The output shows up directly below this cell.
print("Hello from Jupyter!")
print("If you can read this line below the cell, you just ran Python.")
```

### The `In [ ]:` prompt and the output area

Look to the **left** of the code cell you just ran. You'll see a prompt like `In [1]:`.

- `In [ ]:` (empty brackets) → the cell has **not been run** yet.
- `In [*]:` (a star) → the cell is **running right now**.
- `In [1]:` (a number) → the cell **finished**; the number is the *order* it ran in. The first cell you run is `1`, the next is `2`, and so on.

The **output area** is the space directly beneath a code cell where `print(...)` text, tables, and plots appear after you run it.

There's also a small kernel status dot / circle near the top-right of the window. When it's filled/solid the kernel is **busy**; when it's hollow it's **idle** (ready). We'll explain the "kernel" in the next section.

### Markdown cells: raw vs. rendered

This very cell is a **Markdown** cell. Right now you're seeing it rendered (nice heading, bold text). If you **double-click** it, you'll see the raw text with symbols like `##` and `**`. Press Shift + Enter to render it again.

> 💡 A Markdown cell only turns into formatted text **after** you run it with Shift + Enter. If you leave it half-edited, it stays as raw symbols. That's not a bug, you just haven't rendered it yet.

**Tiny Markdown quick-reference:**

### Edit mode vs. command mode (the border color tells you)

Click **inside** a cell to type, and the border turns green. That's **edit mode**: your keystrokes go into the cell.

Press **Esc** (or click the margin just left of the cell), and the border turns blue. That's **command mode**: now single keys are *shortcuts that act on the whole cell*, for example:

- `A` → insert a new cell **above**
- `B` → insert a new cell **below**
- `D` `D` (press D twice) → **delete** the cell
- `M` → turn the cell into **Markdown**; `Y` → turn it back into **code**

## 2. The kernel: one Python brain that remembers

Behind your notebook is a **kernel**, a single Python process that stays alive the whole time the notebook is open. Every code cell you run is handed to that same kernel. Because it's one continuous process, anything you define stays in memory and is available to later cells.

Think of the notebook page as a *formatted history* of commands, and the kernel as the *brain* that actually remembers the results.

Run the next cell to create a variable `x`.

```python
# Define a variable. The kernel will remember this value.
x = 10
print("x is now set to", x)
```

The variable `x` lives in the kernel now. A completely different cell, further down the page, can still see it, with no `import` and no re-defining. Run the next cell:

```python
# We never re-defined x here, but the kernel still remembers it from the cell above.
print("Two cells later, x is still", x)
print("x times 3 is", x * 3)
```

The same memory holds for **functions**. Define one in a cell now:

```python
# Define a function. Like the variable, it stays in the kernel's memory.
def greet(name):
    return "Hi " + name + ", welcome to machine learning!"

print("Function 'greet' is now defined and remembered by the kernel.")
```

```python
# A later cell can call the function we defined above.
print(greet("Ada"))
print(greet("Alan"))
```

> 💡 **Key idea:** the kernel is one Python process that never closes (until you close or restart it). Your cells are a *saved history* you can re-run in any order. That freedom is powerful, and, as the next section shows, it's also the #1 source of confusion for beginners.

## 3. Execution order: the thing that trips everyone up

In a `.py` script, code runs strictly top to bottom. In a notebook, cells run in whatever order **you** press Shift + Enter, which is not necessarily top to bottom. The kernel remembers the *last* value of every variable, no matter which cell set it.

The next cell adds 1 to `n`. First we set `n` to a starting value, then we bump it.

```python
# Start a counter at 1.
n = 1
print("n starts at", n)
```

```python
# This cell adds 1 to n each time you run it.
n = n + 1
print("n is now", n)
```

Now run the cell right above this one three times in a row (click it, then Shift + Enter, Shift + Enter, Shift + Enter). Watch `n` climb: **2, then 3, then 4**.

Nothing about the *code* changed, only *how many times you ran it*. The number in `In [ ]:` on the left keeps going up too, telling you the run order. This is why two people can have the exact same notebook on screen and get different results: they ran the cells a different number of times, or in a different order.

> ⭐ **Golden rule:** when your notebook's numbers stop making sense, restart the kernel and run all the cells to get a clean top-to-bottom run.
> - **Jupyter / JupyterLab:** menu **Kernel → Restart Kernel and Run All Cells…**
> - **Google Colab:** menu **Runtime → Restart session and run all**

**Restart** is not the same as **Clear Outputs**:

- **Restart** wipes the kernel's memory. Every variable and function is forgotten, so you start from a blank brain, and then "Run All" rebuilds everything cleanly.
- **Clear Outputs** only hides the printed results and plots on the page. The kernel still remembers all your variables, so it's purely cosmetic.

## 4. Error clinic + the setup cell

You *will* hit errors, and everyone does. The trick is to **read the traceback from the bottom up**: the very last line names the error type and a short message, and the lines above it show where it happened. Let's meet the two errors this workflow throws most often.

To keep this notebook able to "Restart & Run All" without stopping, we deliberately catch each demo error with `try/except` and print it, instead of letting it halt the page. In real life you'd see the same message in red, and the notebook would stop at that cell.

### Error 1: a `NameError` from using something you never ran

The single most common beginner error: you use a variable, but the cell that *creates* it was never run (you skipped it, or you restarted the kernel and forgot to re-run it). The kernel simply has no idea what that name means.

```python
# We reference a name the kernel has never seen, to show what the error looks like.
try:
    print(total_score)  # 'total_score' was never defined anywhere
except NameError as e:
    print("Got a NameError! Python says:")
    print("   ", e)
    print("\nFix: run the cell that defines 'total_score' first (or Restart & Run All).")
```

The message reads roughly `name 'total_score' is not defined`. The fix is almost always to run the earlier cell that defines it, or do **Restart & Run All** so nothing gets skipped.

### Error 2: a `KeyError` from dropping a column that's already gone

A very common data-cleaning mistake. Say you have a table and you drop a column. If you run that *same* drop cell twice, the second run fails: the column is already gone, so pandas raises a `KeyError`. Let's build a tiny table and reproduce it.

```python
# A tiny throwaway table just to demonstrate the error safely.
import pandas as pd

demo = pd.DataFrame({"keep": [1, 2, 3], "drop_me": [9, 9, 9]})
print("Columns to start:", list(demo.columns))

# First drop works fine.
demo = demo.drop("drop_me", axis=1)
print("After first drop:", list(demo.columns))

# Running the SAME drop again fails, because 'drop_me' is already gone.
try:
    demo = demo.drop("drop_me", axis=1)
except KeyError as e:
    print("\nGot a KeyError on the second drop! Python says:")
    print("   ", e)
```

The fix is one keyword argument: **`errors="ignore"`** tells pandas "if the column isn't there, just skip it." Now the exact same line is safe to run any number of times.

```python
# With errors="ignore", re-running a drop is harmless (no KeyError).
demo = demo.drop("drop_me", axis=1, errors="ignore")   # already gone -> quietly does nothing
demo = demo.drop("drop_me", axis=1, errors="ignore")   # run it again -> still fine
print("Still fine after two more drops. Columns:", list(demo.columns))
```

Writing cells so they're safe to re-run (like `errors="ignore"` above) is a habit that saves you constant grief in notebooks. You'll see it throughout Notebook 2.

### The setup cell

Every notebook in this course starts with **one setup cell** that imports our tools, fixes a random seed so results are reproducible, styles our plots, and defines a `load_csv(...)` helper that finds the data whether you're on Colab or your own machine. **Run it now.** You only run it once per session, and everything after depends on it.

```python
# ============================================================================
# CANONICAL SETUP CELL  (paste verbatim as the first CODE cell in every notebook)
# ============================================================================
# Standard library
import os
import random

# Third-party scientific stack
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# scikit-learn pieces we use across the workflow
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Show plots INSIDE the notebook (this is a Jupyter "magic", not normal Python).
%matplotlib inline

# --- Reproducibility: one seed, used everywhere ------------------------------
# Setting the seed makes "random" steps (shuffling, model training) repeatable,
# so you and your neighbor get the SAME numbers. Change nothing about the data
# and you get identical results every run.
RANDOM_STATE = 42
random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)

# --- Consistent, readable plots ---------------------------------------------
sns.set_theme(style="whitegrid", palette="colorblind")
plt.rcParams["figure.figsize"] = (8, 5)   # a comfortable default size
plt.rcParams["font.size"] = 12             # readable labels

# --- Are we on Google Colab or a local machine? -----------------------------
try:
    import google.colab  # noqa: F401
    ON_COLAB = True
except ImportError:
    ON_COLAB = False

# --- Portable data loader: works in Colab AND local Jupyter ------------------
# It first looks for the bundled CSV next to the repo (data/ folder), and if it
# can't find it (e.g. a bare notebook uploaded to Colab), it downloads the same
# file from the public GitHub repo. Either way you get the identical data.
_DATA_CANDIDATES = ["../data", "data", "day1/data", "../day1/data", "."]
_RAW_BASE = "https://raw.githubusercontent.com/Rhythman/CS_Mines_summer_hs_intern/main/day1/data"

def load_csv(filename):
    """Load a bundled CSV by name, trying local paths first then the repo URL."""
    for folder in _DATA_CANDIDATES:
        path = os.path.join(folder, filename)
        if os.path.exists(path):
            print(f"Loaded '{filename}' from local file: {path}")
            return pd.read_csv(path)
    url = f"{_RAW_BASE}/{filename}"
    print(f"No local copy found - downloading '{filename}' from:\n  {url}")
    return pd.read_csv(url)

# Confirm the environment so students see it worked.
print("Environment:", "Google Colab" if ON_COLAB else "local Jupyter")
print("Versions -> pandas", pd.__version__, "| numpy", np.__version__, "| seaborn", sns.__version__)
import sklearn
print("           scikit-learn", sklearn.__version__)
print("Setup complete. RANDOM_STATE =", RANDOM_STATE)
```

What each part of the setup cell does, one line at a time:

## 5. A brief ML workflow (a clean slice of the Titanic data)

Here's the fun part. The Titanic dataset lists passengers and whether each one survived. We'll build a tiny model that guesses survival from a few passenger facts. This is the exact shape of the workflow you'll practice in Notebook 2, kept short here so you can see all the moving parts at once.

The workflow has six moves: **Load → Peek → Prepare → Split → Train → Evaluate.** One cell each.

> 💡 Quick vocabulary from the lecture: a **feature** is an input clue (like ticket class), and a **label** is the answer we want to predict (here, `Survived`, where 1 = survived and 0 = died). A model learns the pattern from examples where we already know the answer.

### Load: read the passenger table

`load_csv` (from the setup cell) reads the CSV into a pandas **DataFrame**, which is just a spreadsheet-like table held in memory.

```python
# Read the training data into a DataFrame called `train`.
train = load_csv("train.csv")

# `.shape` is (number of rows, number of columns).
print("train has", train.shape[0], "passengers (rows) and", train.shape[1], "columns.")
```

### Peek: look before you leap

`.head()` shows the first 5 rows. Always eyeball your data before modeling.

```python
# Display the first five rows so we can see what the columns look like.
train.head()
```

Each row is one passenger, and each column is a clue about them (age, sex, ticket class `Pclass`, fare, number of siblings/spouses aboard `SibSp`, parents/children aboard `Parch`, and more). The special column **`Survived`** is the answer we want to predict: `1` = survived, `0` = did not.

### Prepare: pick clean features and turn text into numbers

Models do math, so every input must be a **number**. We deliberately choose five columns that have no missing values, so this demo needs no cleanup (Notebook 2 handles messy columns like `Age` and `Cabin`). One of them, `Sex`, is text (`"male"`/`"female"`), so we map it to numbers.

As a deliberate shortcut for this clean, already-complete slice, we encode `Sex` on the whole `train` table here and split afterward. Notebook 2 does it *properly*, splitting first and never overwriting the raw `train` table, so this demo doesn't contradict the golden rule you'll meet later.

```python
# The five clean feature columns we'll feed the model.
features = ["Pclass", "Sex", "Fare", "SibSp", "Parch"]

# `Sex` is text; models need numbers. Map male -> 0, female -> 1.
# (Run the Load cell above again first if you ever re-run just this cell.)
train["Sex"] = train["Sex"].map({"male": 0, "female": 1})

# X = the input features (a table); y = the answer we want to predict (one column).
X = train[features]
y = train["Survived"]

print("X (inputs) shape:", X.shape)
print("y (answers) shape:", y.shape)
X.head()
```

> ⚠️ A pitfall to watch for: `.map(...)` only knows the keys `"male"` and `"female"`. If you re-run this single cell *after* `Sex` is already `0`/`1`, those numbers aren't in the map and quietly become missing values. The safe move is to re-run the **Load** cell first, or just do **Restart & Run All**.

### Split: hold out some passengers to test honestly

We split the data into a **training** part (the model learns from it) and a **validation** part (unseen passengers we grade it on). Testing on data the model already saw would be like grading a student on the exact questions they studied, so the score would lie.

```python
# 80% to train on, 20% held out for honest evaluation.
# stratify=y keeps the same survived/died proportion in both parts.
# random_state=RANDOM_STATE makes the split identical every run.
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print("Training passengers:  ", X_train.shape[0])
print("Validation passengers:", X_val.shape[0])
```

The point of a held-out validation set is to estimate how the model does on passengers it has never seen, and that's the only score that honestly reflects real-world performance.

### Train + Evaluate: the four canonical lines

Almost every scikit-learn model follows the **same four steps**. Learn this rhythm once and it carries over to every model you'll ever use:

1. **Choose** the model (and its settings).
2. **Fit** (train) it on the training data.
3. **Predict** on new data.
4. **Score** those predictions against the true answers.

```python
# 1. CHOOSE: a decision tree, limited to depth 3 so it stays small and readable.
model = DecisionTreeClassifier(max_depth=3, random_state=RANDOM_STATE)

# 2. FIT: let the model learn survival patterns from the TRAINING passengers.
model.fit(X_train, y_train)

# 3. PREDICT: ask the trained model to guess survival for the held-out passengers.
preds = model.predict(X_val)

# 4. EVALUATE: what fraction of those guesses were correct?
acc = accuracy_score(y_val, preds)
print(f"Validation accuracy: {acc:.3f}  ({acc * 100:.1f}% of held-out passengers guessed correctly)")
```

**Is that any good?** Always compare against a dumb baseline. If you ignored every clue and just guessed "everyone died" (the most common outcome), you'd be right about **61.6%** of the time. Our little tree should beat that, and the gap is the value the model actually adds.

```python
# The "everyone died" baseline, for comparison.
baseline = 0.616
print(f"Baseline 'everyone died':   {baseline * 100:.1f}%")
print(f"Our depth-3 decision tree:  {acc * 100:.1f}%")
print("Beats the baseline!" if acc > baseline else "Does not beat the baseline.")
```

### See the rules the model learned

A decision tree is just a flowchart of yes/no questions. Because we kept it to depth 3, we can draw the whole thing and read exactly what it decided.

```python
# Draw the trained decision tree as a flowchart of yes/no rules.
plt.figure(figsize=(16, 8))
plot_tree(
    model,
    feature_names=features,              # label splits with our column names
    class_names=["Died", "Survived"],    # label the outcomes in plain English
    filled=True,                          # color boxes by predicted class
    rounded=True,
    fontsize=10,
)
plt.title("The rules the computer wrote from the data (read top to bottom)")
plt.tight_layout()
plt.show()
```

> 🔎 **What to notice in the tree**
> - Read top to bottom. Each box asks one yes/no question about a feature, with left = True and right = False.
> - The very first split is on `Sex`. The model discovered on its own that sex is the strongest single clue to survival.
> - Boxes shaded toward one color are more "sure." The `class = Died` vs `class = Survived` line at the bottom is the tree's final guess for a passenger who lands there.
>
> Nobody hand-wrote these rules. The model *learned* them from the training passengers, and that's exactly why we keep the tree shallow at `max_depth=3`: a small tree we can actually read and trust.
